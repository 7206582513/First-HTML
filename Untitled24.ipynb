```python
gender_product_preferences = df.groupby(['Gender', 'ProductPitched']).size().unstack(fill_value=0)
gender_product_preferences
```

```python
trip_purchase_relationship = df.groupby('NumberOfTrips')['ProductPitched'].value_counts(normalize=True).unstack()
```

```python
passport_trip_correlation = df['Passport'].corr(df['NumberOfTrips'])
```

```python
passport_trip_correlation
```

```python
customers_with_children = df[df['NumberOfTrips'] > 0]
grouped_by_children = customers_with_children.groupby('NumberOfTrips')
average_purchases_by_children = grouped_by_children['ProductPitched'].value_counts().unstack().fillna(0).mean(axis=1)
```

```python
gender_preference=df.groupby('Gender')['PitchSatisfactionScore'].value_counts().unstack(fill_value=0)
```

```python
purchase_by_martial_status=df.groupby('MaritalStatus')['ProductPitched'].value_counts().unstack().fillna(0)
purchase_by_martial_status
```

```python
purchase_percentage_by_marital_status = purchase_by_martial_status.div(purchase_by_martial_status.sum(axis=1), axis=0) * 100
purchase_percentage_by_marital_status
```

```python
no_of_children_by_product_interest=df.groupby('NumberOfChildrenVisiting')['Designation'].value_counts().unstack().fillna(0)
no_of_children_by_product_interest
```

```python
no_of_own_car_by_Pitch_duration=df.groupby('OwnCar')['NumberOfTrips'].value_counts().unstack().fillna(0)
no_of_own_car_by_Pitch_duration
```
